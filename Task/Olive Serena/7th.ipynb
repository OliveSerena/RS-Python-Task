```python
def fact(N):
  for i in range(2,N+1) :
    if(N%2==0):
      if isprime(i):
        return i
  return N
```

```python
def isprime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```
