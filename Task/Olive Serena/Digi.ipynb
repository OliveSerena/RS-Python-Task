```python
a=input("Enter a number : ")
l= len(a)
digits=[]
for i in a :
  digits.append(int(i))
print(f"Original Digits = {digits}")
if digits[l-1] != 9:
  digits[l-1] = (digits[l-1]+1)
if digits[l-1] == 9:
  digits[l-1]=1
  digits.append(0)
  l=l+1
print(f"Digits = {digits}")
```
